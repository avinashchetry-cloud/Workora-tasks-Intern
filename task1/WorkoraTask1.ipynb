### `Filled the missing values of director and cast, country with unknown because the missing values were big in number and we couldn't afford to lose majority of the data!`

```python
df = pd.read_csv('netflix_titles.csv')
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df.describe()
```

```python
print(df.shape)
print(df.isnull().sum())
```

```python
(df.isna().sum()/len(df))*100
```

```python
df['director'].isna().sum()
```

```python
df['director'].fillna('Unknown', inplace=True)
df['director'].isna().sum()
```

```python
df['cast'].isna().sum()
```

```python
df['cast'].fillna('Unknown', inplace=True)
df['cast'].isna().sum()
```

```python
df['country'].isna().sum()
```

```python
df['country'].fillna('Unknown', inplace=True)
df['country'].isna().sum()
```

```python
df['date_added'].isna().sum()
```

```python
df.dropna(subset=['date_added'], inplace=True)
df['date_added'].isna().sum()
```

```python
df['rating'].isna().sum()
```

```python
df.dropna(subset=['rating'], inplace=True)
df['rating'].isna().sum()
```

```python
df['duration'].isna().sum()
```

```python
df.dropna(subset=['duration'], inplace=True)
df['duration'].isna().sum()
```

```python
df.shape
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
df.dtypes
```

```python
df['date_added']
```

```python
df_strings = df.select_dtypes(include='object') # selecting object dtype columns
df_strings.head()  
```

```python
def stripObjCols(strObj):
  return strObj.strip() # to strip the string form each obj type columns
```

```python
for cols in df_strings.columns:
  df[cols] = df_strings[cols].apply(stripObjCols)
```

```python
df["date_added"] = pd.to_datetime(df["date_added"], format="%B %d, %Y")
```

```python
df.dtypes
```

```python
(df.isna().sum()/len(df))*100
```

```python
df["type"] = df["type"].str.title() # converting to generic title format
```

```python
df["country"] = df["country"].str.title()
```

```python
print(df.shape)
print(df.isnull().sum())
```

```python
df.to_csv('cleaned.csv', index=False) # exporting the dataframe to csv file
```
